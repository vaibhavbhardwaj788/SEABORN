```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #seaborn is based graph 
sns.set(color_codes=True)#Adds nice background to graph 
%matplotlib inline
#Tell python to actually display the graphs
```

```python
auto=pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['highway_mpg'])
```

```python
sns.displot(auto['city_mpg'],kde=True,rug=True)
```

```python
sns.displot(auto['city_mpg'],kde=False,rug=True)
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower']) #jointplot creates scatter plot of two varibles along with histogram
```

```python
#Hexa Plot
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="hex")
```

```python
# 2D curve 
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="kde")
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
```

```python
#pair wise plot
sns.stripplot(auto['fuel_type'],auto['horsepower'],jitter=True)
```

```python
#plotting with categorial data, adust position using jitters
sns.stripplot(auto['fuel_type'],auto['horsepower'],jitter=False)
```

```python
#Swarmplot
sns.swarmplot(auto['fuel_type'],auto['horsepower'])
```

```python
#Box Plot - this kind of box plot shows 3 quartiles values along with extreme value.
#The "Wiskers" extend to point that lie within 1.5 IQRS of the lower and upper quartile.-
sns.boxplot(auto['number_of_doors'],auto['horsepower'])
```

```python
sns.boxplot(auto['number_of_doors'],auto['horsepower'],hue=auto['fuel_type'])
```

```python
#Bar plot using (HUE)
sns.barplot(auto['body_style'],auto['horsepower'],hue=auto['fuel_type'])
```

```python
#count plot
sns.countplot(auto['body_style'])
```

```python
sns.catplot(x="fuel_type",
           y="horsepower",
           hue="number_of_doors",
           col="drive_wheels",
           data=auto,kind="violin")
```

```python
#Point plot
sns.pointplot(auto['body_style'],auto['horsepower'],hue=auto['number_of_doors'])
```

```python
#Vairious types of inputs :(point ,bar,count,box,violin,strip)
```

```python
#Linear regression plot
sns.lmplot(y="horsepower",x="engine_size",data=auto)
```

```python
sns.lmplot(y="horsepower",x="engine_size",hue="fuel_type",data=auto)
```
